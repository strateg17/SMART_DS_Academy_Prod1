```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("full_data.csv")
df
```

```python
df.columns
```

```python
df[df["polarity"]=='positive']['product_category_name'].value_counts().head(10)
```

```python
data = df[df["polarity"]=='positive']['product_category_name'].value_counts().head(10)
plt.figure(figsize=(12, 8))
x = list(data.index)
y = list(data.values)
x.reverse()
y.reverse()

plt.title("Product categories with most positive reviews")
plt.ylabel("Product categories")
plt.xlabel("Reviews")

plt.barh(x, y)
```

```python
df.groupby(['product_category_name']).agg({'review_model_score': 'mean'}).sort_values(by='review_model_score').head(10)['review_model_score'].values
```

```python
data = df.groupby(['product_category_name']).agg({'review_model_score': 'mean'}).sort_values(by='review_model_score').head(10)['review_model_score']
plt.figure(figsize=(12, 8))
x = list(data.index)
y = list(data.values)
x.reverse()
y.reverse()

plt.title("Product categories with lowest mean")
plt.ylabel("Product categories")
plt.xlabel("Reviews")

plt.barh(x, y)
```

```python
data = df.groupby(['product_category_name']).agg({'review_model_score': 'mean'}).sort_values(by='review_model_score').tail(10)['review_model_score']
plt.figure(figsize=(12, 8))
x = list(data.index)
y = list(data.values)
x.reverse()
y.reverse()

plt.title("Product categories with highest mean")
plt.ylabel("Product categories")
plt.xlabel("Reviews")

plt.barh(x, y)
```
