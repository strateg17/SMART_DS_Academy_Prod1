```python
from polyglot.text import Text
```

```python
df = pd.read_csv('order_reviews.csv')
df = df.drop(['review_id','order_id','review_comment_title','review_creation_date','review_answer_timestamp'],axis=1)
df.dropna(subset = ["review_comment_message"], inplace=True)
df
```

```python
df_final = df[:20000]
df_final
```

```python
def review_polyglot_score(text):
    try:
        text = Text(text)
        score = 0
        for i in text.words:
            score+=i.polarity
        return score/len(text.words)
    except:
        return 0.0
    
```

```python
st = " "
review_polyglot_score(st)
```

```python
df_final['review_polyglot_score']=df_final['review_comment_message'].apply(lambda x: review_polyglot_score(x))
df_final
```
